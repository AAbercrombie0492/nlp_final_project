```python
from __future__ import division

import graphlab as gl
import pandas as pd
import pyLDAvis
import pyLDAvis.graphlab
import pickle
from sklearn.feature_extraction.text import CountVectorizer
import re
from collections import Counter

pyLDAvis.enable_notebook()
```

```python
#Load Data
balanced_classes = pd.read_csv('balanced_classes.csv')
travel= pd.read_csv('travel.csv')
biology= pd.read_csv('biology.csv')
```

```python
#Construct dictionary of word counts for each row in each data set
balanced_classes['bow'] = balanced_classes.ttl_ctxt.apply(lambda x: dict(Counter(x.split())))
travel['bow'] = travel.ttl_ctxt.apply(lambda x: dict(Counter(x.split())))
biology['bow'] = biology.ttl_ctxt.apply(lambda x: dict(Counter(x.split())))
```

```python
#Graphlab requires special s_frame objects
balanced_classes_sf = gl.load_sframe(balanced_classes)
balanced_classes_bows = balanced_classes_sf['bow']

travel_sf = gl.load_sframe(travel)
travel_bows = travel_sf['bow']

biology_sf = gl.load_sframe(biology)
biology_bows = biology_sf['bow']
```

```python
balanced_classes_topic_model = gl.topic_model.create(balanced_classes_bows, num_topics=6, num_iterations=200)
```

# Balanced Classes

```python
pyLDAvis.graphlab.prepare(balanced_classes_topic_model, balanced_classes_bows)
```

```python
travel_topic_model = gl.topic_model.create(travel_bows, num_topics=9, num_iterations=200)
```

# Travel

```python
pyLDAvis.graphlab.prepare(travel_topic_model, travel_bows)
```

```python
bio_topic_model = gl.topic_model.create(biology_bows, num_topics=5, num_iterations=200)
```

# Biology

```python
pyLDAvis.graphlab.prepare(bio_topic_model, biology_bows)
```
